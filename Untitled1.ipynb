range() 함수와 유사한 frange()함수를 다음 조건을 참고하여 만드시오.
- 1) frange()의 인자 구성은 다음과 같이 range와 동일 하지만 각 인수들은 음수를 받지 않는다고 가정한다.
>range(stop)  
>range(start, stop[, step])

- 2) frange() 함수의 인자에 대한 기본 시작(start) 값은 0.0 이고, 기본 단계(step) 값은 0.1이다.
- 3) frange 사용 예
 - 3-1) frange(0.5)
   - [0.0, 0.1, 0.2, 0.3, 0.4]
 - 3-2) frange(1.0, 2.0)
   - [1.0, 1.1, 1.2, 1.3, 1.4, 1.5, 1.6, 1.7, 1.8, 1.9]
 - 3-3) frange(2.2, 4.0, 0.5)
   - [2.2, 2.7, 3.2, 3.7]

```python
def frange(*args):
    if len(args) == 1:
        if args[0] < 0 : #음수 들어왔을 때 예외처리
            return
        stop = args[0]
        start = 0.0
        index = start
        step = 0.1
        tmp_list = []
        while True:
            tmp_list.append(index)
            index += step
            if index >= stop:
                break
        return tmp_list
    elif len(args) == 2:
        if args[0] < 0 :
            return
        if args[1] < 0 : #음수 들어왔을 때 예외처리
            return
        start = args[0]
        stop = args[1]
        index = start
        step = 0.1
        tmp_list = []
        while True:
            tmp_list.append(index)
            index += step
            if index >= stop:
                break
        return tmp_list
    elif len(args) == 3:
        if args[0] < 0 :
            return
        if args[1] < 0 : 
            return 
        if args[2] < 0 :  #음수 들어왔을 때 예외처리
            return 
        start = args[0]
        stop = args[1]
        index = start
        step = args[2]
        tmp_list = []
        while True:
            tmp_list.append(index)
            index += step
            if index >= stop:
                break
        return tmp_list
    else:
        print "error : no match number of arguments"

print frange(0.5)
print frange(1.0, 2.0)
print frange(2.2, 4.0, 0.5)
```

4번 문제  
가변인수를 받는 함수 sum()을 다음과 같은 조건을 참고하여 구현하시오  
- sum() 사용 예
 - sum()
   - 0
 - sum(1, 2)
   - 3
 - sum(1, 2, 3, 4, 5)
   - 15
 - sum(1, 5, 7, 2, -10)
   - 5

```python
def sum(*args):
    tmp = 0
    for i in range(len(args)):
        tmp += args[i]
    return tmp

print sum(0)
print sum(1, 2)
print sum(1, 2, 3, 4, 5)
print sum(1, 5, 7, 2, -10)
```

5번 문제  
여러 단어로 이루어진 문자열을 입력받아 각 단어의 첫 글자로 이루어진 단어를 대문자로 출력하는 myinitial() 함수를 다음 조건을 참고하여 작성하시오.  
- 1) 다음에 제시되는 함수들을 모두 이용해야 한다.
 - split
 - map
 - join
- 2) myinitial() 함수 사용 예
 - myinitial("as soon as possible")
   - ASAP

```python
def myinitial(*args):
    tmp = args[0]
    tmp_list = tmp.split()
    for i in range(len(tmp_list)):
        tmp_list[i] = tmp_list[i].capitalize()    #첫글자를 대문자로 변환
    tmp = []
    map(lambda x: tmp.append(tmp_list[x][0]), range(len(tmp_list)))
    return ''.join(tmp)    
    
myinitial("as soon as possible")
```

6번 문제  
음이 아닌 정수 n를 입력받으면 n! (factorial)을 계산하는 myfact() 함수를 재귀적 함수로 구현하시오.  
- [옵션] 참을 수 있는 정도 만큼의 수행시간을 직접 기다려 보면서 n을 늘려보도록 합시다.

```python
def myfact(a):
    if a == 1:
        return 1
    return a * myfact(a-1)

print myfact(976)
```

####프로젝트 오일러 문제 10

10 이하의 소수를 모두 더하면 2 + 3 + 5 + 7 = 17 이 됩니다.  
이백만(2,000,000) 이하 소수의 합은 얼마입니까?

```python
#에라토스테네스의 체를 이용한 방법
setting = input()
primes = range(1, setting+1)
primes[0] = 0
index = 1
while index < setting:
    if primes[index]:
        i = 2
        while primes[index]*i <= setting:
            primes[(index+1)*i-1] = 0
            i = i + 1
    index = index + 1
print sum(primes)    
```

양의 정수 n에 대하여, 다음과 같은 계산 과정을 반복하기로 합니다.  
$n -> n/2 (n이 짝수일 때)$  
$n -> 3n+1 (n이 홀수일 때)$  
13에 대하여 위의 규칙을 적용해보면 아래처럼 10번의 과정을 통해 1이 됩니다.  
13 -> 40 -> 20 -> 10 -> 5 -> 16 -> 8 -> 4 -> 2 -> 1  
아직 증명은 되지 않았지만, 이런 과정을 거치면 어떤 수로 시작해도 마지막에는 1로 끝나리라 생각됩니다.  
그러면, 백만(1,000,000) 이하의 수로 시작했을 때 1까지 도달하는데 가장 긴 과정을 거치는 숫자는 얼마입니까?


```python
hseqdict = {} # {갯수: 숫자}
for i in range(1, 1000001):
    hseq = [i]
    while hseq[-1] != 1:
        if hseq[-1] % 2 == 0:
            hseq.append(hseq[-1] / 2)
        else:
            hseq.append(3 * hseq[-1] + 1)
    hseqdict[len(hseq)] = i

print (hseqdict[max(hseqdict.keys())])
```

1) 다음 사이트에서 제시되는 영어 불용어를 참고하여 이전 숙제에서 구성했던 단어 사전에서 영어 불용어들을 모두 제거하는 코드를 추가하시오.

영어 불용어: [ 'a', 'and', 'are', 'as', 'at', 'be', 'but', 'by', 'for', 'if', 'in', 'into', 'is', 'it', 'no', 'not', 'of', 'on', 'or', 's', 'such', 't', 'that', 'the', 'their', 'then', 'there', 'these', 'they', 'this', 'to', 'was', 'will', 'with']  
2) 각 URL로 지정된 웹페이지의 HTML 소스를 파일로 저장하시오.  
URL이 http://URL 이라면 파일명은 URL.html 이다.  
예: URL이 http://www.cnn.com 이라면 파일명은 www.cnn.com.html 이다.  
3) 단어의 출현빈도가 담긴 사전 객체를 위 HTML 소스 파일과 동일한 폴더에 파일로 저장하시오.  
파일입출력 (E-learning 13주차) 마지막에 학습한 pickle 모듈을 활용하시오.  
URL이 http://URL 이라면 사전 객체를 담고 있는 파일명은 URL.words.frequency 이다.  
예: URL이 http://www.cnn.com 이라면 파일명은 www.cnn.com.words.frequency 이다.  
4) 최소 5개 이상의 영어 웹 사이트에 대한 HTML 소스 파일과 단어 출현빈도 파일을 저장하시오.  
즉, 총 10개의 파일을 동일한 폴더에 생성하시오.  
[주의] 영어 웹사이트 URL로만 5개 이상  
5) 위 문제에서 저장한 모든 words.frequency 파일들을 객체로 다시 로드하여 본인이 저장하여 분석한 사이트들에서 가장 많이 출현한 단어 3개를 뽑아 제시하시오.  
반드시 pickle 모듈로 저장한 5개 이상의 words.frequency를 다시 5개 이상의 사전 객체로 로드 하는 코드가 추가되어야 함  

```python
import urllib2
import string
import json
import pickle

def func(source):
    src_max = len(source)
    i = 0
    mode = True
    result = []
    tmp = ""
    res_string = []
    while i+1 < src_max:
        if source[i] == '<':
            mode = False           #태그가 시작되면 문자를 저장하지 않음
            result.append(tmp)     #지금까지 저장한 글자들을 result 리스트에 저장
            tmp=""                 #버퍼 비움
        elif source[i] == '>':
            mode = True            #태그가 끝났으므로 문자를 저장할 준비를 함
            tmp = ""               #주석에 대해서 예외처리
        elif mode:
            tmp += source[i]       #현재 문자를 저장해야할 시점이므로 버퍼에 저장
        i = i + 1       #다음 글자로 이동
    for i in range(0, len(result)):      #구두문자 제거
        str = result[i]
        result[i] = str.translate(None, string.punctuation)
    for i in range(0, len(result)):
        result[i] = result[i].split()    #공백을 split으로 없앰
    for i in range(0, len(result)):
        for j in range(0, len(result[i])):
            res_string.append(unicode(result[i][j], 'utf-8'))
            #split으로 쪼갠 단어들은 이차원 리스트에 존재하므로 하나하나 다시 뽑아냄
            #이차원 원소는 현재 모두 유효한 값만을 지니고 있음
    # 불용어 제거 부분
    unable_list = [ u'a', u'and', u'are', u'as', u'at', u'be', u'but', u'by', u'for', u'if', u'in', u'into', u'is', u'it', u'no', u'not', u'of', u'on', u'or', u's', u'such', u't', u'that', u'the', u'their', u'then', u'there', u'these', u'they', u'this', u'to', u'was', u'will', u'with']
    res_string_new = []
    for i in range(len(res_string)):
        if res_string[i] in unable_list:
            continue
        else:
            res_string_new.append(res_string[i])
    # 불용어 제거 부분 끝
    D = {}
    for i in range(len(res_string_new)):
        if res_string_new[i] in D:
            D[res_string_new[i]] += 1
        else:
            D[res_string_new[i]] = 1
    return D

source = urllib2.urlopen("https://www.ted.com").read()
source2 = urllib2.urlopen("https://www.apple.com").read()
source3 = urllib2.urlopen("http://www.yale.edu/").read()
source4 = urllib2.urlopen("https://www.qualcomm.com/").read()
source5 = urllib2.urlopen("http://web.mit.edu/").read()
#Source를 html로 저장
f = file("www.ted.com.html", "w")
f.write(source)
f.close()
f = file("www.apple.com.html", 'w')
f.write(source2)
f.close()
f = file("www.yale.edu.html", 'w')
f.write(source3)
f.close()
f = file("www.qualcomm.com.html", 'w')
f.write(source4)
f.close()
f = file("web.mit.edu.html", 'w')
f.write(source5)
f.close()
#Source를 html로 저장 끝
ted = func(source)
apple = func(source2)
yale = func(source3)
qualcomm = func(source4)
mit = func(source5)
#pickle로 저장
f = file("www.ted.com.words.frequency", "w")
pickle.dump(ted, f)
f.close()
f = file("www.apple.com.words.frequency", "w")
pickle.dump(ted, f)
f.close()
f = file("www.yale.edu.words.frequency", "w")
pickle.dump(ted, f)
f.close()
f = file("www.qualcomm.com.words.frequency", "w")
pickle.dump(ted, f)
f.close()
f = file("web.mit.edu.words.frequency", "w")
pickle.dump(ted, f)
f.close()
#pickle로 저장 끝
#pickle load test
f = file("www.ted.com.words.frequency", "r")
t = pickle.load(f)
print json.dumps(t, ensure_ascii=False, indent=4)
f.close()
f = file("www.apple.com.words.frequency", "r")
t = pickle.load(f)
print json.dumps(t, ensure_ascii=False, indent=4)
f.close()
f = file("www.yale.edu.words.frequency", "r")
t = pickle.load(f)
print json.dumps(t, ensure_ascii=False, indent=4)
f.close()
f = file("www.qualcomm.com.words.frequency", "r")
t = pickle.load(f)
print json.dumps(t, ensure_ascii=False, indent=4)
f.close()
f = file("web.mit.edu.words.frequency", "r")
t = pickle.load(f)
print json.dumps(t, ensure_ascii=False, indent=4)
f.close()
#pickle load test 끝
```
